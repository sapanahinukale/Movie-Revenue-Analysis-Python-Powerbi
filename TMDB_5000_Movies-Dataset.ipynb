status - "Released" or "Rumored".

# Data Cleaning

```python
# checking null values
tmdb_data.isnull().sum()
```

```python
# Visualizing missing values in the tmdb_data DataFrame
sns.heatmap(tmdb_data.isnull())
```

```python
# Drop columns not required further for analysis
tmdb_data.drop(['homepage', 'keywords', 'original_language', 'original_title', 'tagline', 'overview', 'spoken_languages', 'status'], axis=1, inplace = True) # inplace = true means it modifies the original dataframe and changes are reflects in the original object.
tmdb_data.head(2)
```

```python
# Check the null counts again of tmdb_data
tmdb_data.isnull().sum()
```

```python
# Fill missing runtime with the median value (recommended for runtime)
tmdb_data['runtime'].fillna(tmdb_data['runtime'].median(), inplace=True)
```

```python
tmdb_data.dropna()
```

```python
# convert the data in the 'release_date' column to datetime format.
tmdb_data['release_date'] = pd.to_datetime(tmdb_data['release_date'])
tmdb_data.dtypes
```

```python
import numpy as np
# Plot boxplots for all numerical columns to visualize outliers
numerical_cols = tmdb_data.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)  # Adjust the grid size based on number of columns
    sns.boxplot(data=tmdb_data, x=col)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Assuming tmdb_data is your DataFrame
numerical_cols = tmdb_data.select_dtypes(include=[np.number]).columns

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numerical column
Q1 = tmdb_data[numerical_cols].quantile(0.25)
Q3 = tmdb_data[numerical_cols].quantile(0.75)

# Calculate the Interquartile Range (IQR) for each column
IQR = Q3 - Q1

# Define a condition to filter out the outliers (values outside 1.5*IQR range)
condition = ~((tmdb_data[numerical_cols] < (Q1 - 1.5 * IQR)) | (tmdb_data[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Filter out rows that contain outliers
tmdb_data_no_outliers = tmdb_data[condition]

# Print the shape of the original and cleaned dataset
print(f"Original Data: {tmdb_data.shape}")
print(f"Data after removing outliers: {tmdb_data_no_outliers.shape}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize boxplots before removing outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=tmdb_data, x=col)
    plt.title(f'Boxplot of {col} - Original Data')
plt.tight_layout()
plt.show()

# Visualize boxplots after removing outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=tmdb_data_no_outliers, x=col)
    plt.title(f'Boxplot of {col} - Data without Outliers')
plt.tight_layout()
plt.show()
```

```python
import json

# Convert the 'genres' column to a list of dictionaries using json.loads
tmdb_data['genres'] = tmdb_data['genres'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
```

```python
# Extract genre names into a new 'genre_names' column
tmdb_data['genre_names'] = tmdb_data['genres'].apply(lambda x: [genre['name'] for genre in x] if isinstance(x, list) else [])

# Now print the result
print(tmdb_data[['title', 'genre_names']])
```

```python
# This column has data in zip format which can't be analysed further through present knowledge.these columns are of no use and we need to drop these columns.
tmdb_data.drop(['production_companies', 'production_countries'], axis=1, inplace=True)
tmdb_data.head(2)
```

# Visualization

```python
# Plot a histogram of movie runtime
plt.hist(tmdb_data['runtime'], bins=15, edgecolor='black')  # Create a histogram with 15 bins and black edges
plt.xlabel('Runtime')  # Label the x-axis as 'Runtime'
plt.ylabel('Frequency')  # Label the y-axis as 'Frequency'
plt.title('Distribution of Movie Runtimes')  # Set the title of the plot
plt.show()  # Display the plot
```

```python
# Visualize the distribution of original languages
language_counts = tmdb_data['runtime'].value_counts().head(10)
sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')
plt.xlabel('runtime')
plt.ylabel('Count')
plt.title('Top 10 Original titles in Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting a scatter plot of budget vs. revenue using Seaborn
sns.scatterplot(x='budget', y='revenue',data=tmdb_data)  # Scatter plot using Seaborn

# Adding labels and title to the plot
plt.xlabel('Budget')  # Label for the x-axis as 'budget'
plt.ylabel('Revenue')  # Label for the y-axis as 'revenue'
plt.title('Budget vs. Revenue')  # Title of the plot

# Displaying the plot
plt.show()  # Show the plot
```

```python
# Plot a line plot of average revenue over release years
# Grouping the data by release year and calculating the average revenue
avg_revenue_by_year = tmdb_data.groupby(tmdb_data['release_date'].dt.year)['revenue'].mean()

# Plotting the line plot
plt.plot(avg_revenue_by_year.index, avg_revenue_by_year.values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Release Year')
plt.ylabel('Average Revenue')
plt.title('Average Revenue Over Release Years')

# Displaying the plot
plt.show()
```
